# Предварительный анализ данных

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
url = 'https://raw.githubusercontent.com/vvwsL/int_25/4c29b4ec5cd0536f5b19786981c5baa8c627af16/T_games_dataset.csv'
df = pd.read_csv(url)

# Преобразование даты
df['order_day'] = pd.to_datetime(df['order_day'])

# Основная информация
display(df.info())
display(df.describe(include='all'))
display(df.head())
```

# Решение задачи 1: Анализ портретов покупателей видеоигр

## Как игроки отличаются от других клиентов Т-Банка?

```python
# Создаем фигуру с 3 подграфиками
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Анализ демографии покупателей игр', fontsize=16)

# 1. Распределение возраста и пола
sns.histplot(data=df, x='age', hue='gender_cd', kde=True, multiple='stack', ax=axes[0])
axes[0].set_title('Распределение по возрасту и полу')
axes[0].set_xlabel('Возраст')
axes[0].set_ylabel('Количество клиентов')

# 2. Boxplot дохода по полу
sns.boxplot(data=df, x='gender_cd', y='monthly_income_amt', ax=axes[1])
axes[1].set_title('Распределение дохода по полу')
axes[1].set_xlabel('Пол')
axes[1].set_ylabel('Месячный доход')

# 3. Распределение по уровню образования
edu_order = ['SCH', 'UGR', 'GRD', 'PGR', 'ACD']
sns.countplot(data=df, x='education_level', order=edu_order, ax=axes[2])
axes[2].set_title('Распределение по уровню образования')
axes[2].set_xlabel('Уровень образования')
axes[2].set_ylabel('Количество клиентов')

plt.tight_layout()
plt.show()

# Дополнительный анализ по возрасту и доходу
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='monthly_income_amt', hue='gender_cd', alpha=0.6)
plt.title('Зависимость дохода от возраста')
plt.xlabel('Возраст')
plt.ylabel('Месячный доход')
plt.grid(True)
plt.show()
```

Согласно визуализациям распределения возраста и пола, игроки — преимущественно мужчины 25–35 лет со среднем доходом. Можно заметить, что основная масса клиентов это люди, находящиеся на одном из 3 уровней обучения: SCH - начальное, среднее; UGR - неполное высшее; GRD - высшее. Это видно на соответствующих гистограммах и boxplot по доходу.

## Можно ли выделить группы игр с разной аудиторией?

```python
from matplotlib import pyplot as plt

# Группировка данных
category_stats = df.groupby('category_name').agg({
    'age': 'mean',
    'monthly_income_amt': 'mean',
    'gender_cd': lambda x: (x == 'M').mean()  # доля мужчин
}).sort_values('age')

# Создаем фигуру
fig, axes = plt.subplots(1, 3, figsize=(20, 8))
fig.suptitle('Анализ аудитории по игровым жанрам', y=1.05)

# 1. Средний возраст
category_stats['age'].plot(kind='barh', ax=axes[0], color='#1f77b4')
axes[0].set_title('Средний возраст покупателей')
axes[0].set_xlabel('Возраст')
axes[0].spines[['top', 'right']].set_visible(False)

# 2. Средний доход
category_stats['monthly_income_amt'].plot(kind='barh', ax=axes[1], color='#ff7f0e')
axes[1].set_title('Средний доход покупателей')
axes[1].set_xlabel('Доход')
axes[1].spines[['top', 'right']].set_visible(False)

# 3. Гендерное распределение
category_stats['gender_cd'].plot(kind='barh', ax=axes[2], color='#2ca02c')
axes[2].set_title('Доля мужчин среди покупателей')
axes[2].set_xlabel('Доля мужчин')
axes[2].set_xlim(0, 1)
axes[2].spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```

Например можно выделить следующие категории:


1.   Стратегии и RPG чаще покупают мужчины 25–35 лет.
2.   Казуальные игры чаще выбирают женщины и старшая аудитория.
3.   Хоррор и гонки — выбирают люди моложе и с доход выше среднего.

## Справедливы ли выводы во времени?

```python

# Динамика покупок по годам
df['year'] = df['order_day'].dt.year
yearly_data = df.groupby('year').agg({'good_id': 'count', 'good_price': 'sum'}).rename(columns={'good_id': 'Кол-во покупок', 'good_price': 'Выручка'})

yearly_data.plot(kind='bar', subplots=True, figsize=(12, 6), title=['Количество покупок', 'Выручка'])
plt.tight_layout()
plt.show()



```

Временной анализ показывает стабильные паттерны. Поведение клиентов не меняется принципиально между 2022 и 2023 годом.

# Решение задачи 2: Анализ ценовой эластичности

## Исследование цен и объемов

```python

# Расчет средней цены и спроса на игру
game_stats = df.groupby('good_id').agg({'good_price': 'mean', 'good_cnt': 'sum'}).reset_index()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=game_stats, x='good_price', y='good_cnt')
plt.title('Зависимость между ценой и количеством покупок')
plt.xlabel('Средняя цена игры')
plt.ylabel('Спрос (объем продаж)')
plt.show()
```

## Популярность в Steam

```python
# Анализ влияния популярности в Steam на продажи
df_filtered = df[df['steam_popularity_score'] > 0]
popularity_stats = df_filtered.groupby('good_id').agg({
    'steam_popularity_score': 'mean',
    'good_price': 'mean',
    'good_cnt': 'sum'
}).reset_index()

# Инверсия популярности
popularity_stats['inv_popularity'] = 5001 - popularity_stats['steam_popularity_score']

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=popularity_stats, x='inv_popularity', y='good_cnt')
plt.title('Популярность в Steam vs Продажи')
plt.xlabel('Популярность (чем выше, тем популярнее)')
plt.ylabel('Объем продаж')
plt.grid(True)
plt.show()
```

## Рекомендации

Исследование цен и объемов:
1.   Недорогие игры чаще покупаются, что может говорить об эластичном спросе.
2.   Игры с высокой популярностью в Steam покупаются чаще даже при высокой цене.
